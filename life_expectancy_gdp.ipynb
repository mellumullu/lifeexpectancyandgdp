My goal for this project is to answer these questions:  

1)  Has life expectancy increased over time in the six nations?
2)  Has GDP increased over time in the six nations?
3)  Is there a correlation between GDP and life expectancy of a country?
4)  What is the average life expectancy in these nations?
5)  What is the distribution of that life expectancy?

```python
# Load the data and fix the current working directory:

import os
import pandas as pd

print("Current Working Directory:", os.getcwd())

# Define the new path
new_path = '/Users/merlinkontus/Desktop/Life-Expectancy-and-GDP-starter'  

# Change the working directory
try:
    os.chdir(new_path)
    print("Changed Working Directory to:", os.getcwd())
except FileNotFoundError:
    print(f"Directory not found: {new_path}")

# Define the path to the CSV file
file_path = 'all_data.csv'  

# Load the CSV file into a DataFrame
try:
    data = pd.read_csv(file_path)
    print(data.head())
except FileNotFoundError:
    print(f"File not found: {file_path}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

1)  Has life expectancy increased over time in the six nations?

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")
df = data

plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country', marker='o')

# Adding title and labels
plt.title('Life Expectancy Over Time in Six Nations')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')

# Show plot
plt.show()
```

2)  Has GDP increased over time in the six nations?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country', marker='o')

# Set logarithmic scale for y-axis
plt.yscale('log')

# Adjust y-axis limits to avoid compressing low values too much
plt.ylim(bottom=100000000)  # Set a minimum value for y-axis

# Adding title and labels
plt.title('GDP Over Time in Six Nations (Logarithmic Scale)')
plt.xlabel('Year')
plt.ylabel('GDP (in billion USD)')

# Show plot
plt.legend(loc='lower right')  # Adjust legend position as needed
plt.grid(True)  # Add grid for better readability
plt.tight_layout()  # Ensure tight layout
plt.show()
```

3)  Is there a correlation between GDP and life expectancy of a country?

```python
# Drop rows with negative or non-numeric GDP values

plt.figure(figsize=(17, 10))
plt.scatter(df['GDP'], df['Life expectancy at birth (years)'], marker='o', s=100, alpha=0.4, edgecolors='black')

# Adding title and labels
plt.title('Relationship Between GDP and Life Expectancy')
plt.xlabel('GDP (in billion USD)')
plt.ylabel('Life Expectancy (years)')


# Customize gridlines
plt.grid(True, linestyle='--', alpha=0.5)

# Adjusting spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show plot
plt.tight_layout()
plt.show()
```

4)  What is the average life expectancy in these nations?

```python
average_life_expectancy = df['Life expectancy at birth (years)'].mean()

plt.figure(figsize=(10, 6))
plt.bar(df['Country'], df['Life expectancy at birth (years)'], color='skyblue', label='Life Expectancy')

# Plotting the average line
plt.axhline(y=average_life_expectancy, color='red', linestyle='--', label=f'Average: {average_life_expectancy:.2f}')

# Adding title and labels
plt.title('Average Life Expectancy Across Nations')
plt.xlabel('Country')
plt.ylabel('Life Expectancy (years)')
plt.legend()

plt.ylim(55, None)

# Customize gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```

5)  What is the distribution of that life expectancy?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Life expectancy at birth (years)'], bins=20, kde=True)

plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy (years)')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Conclusion:

Life expectancy has increased in all 6 countries. Zimbabwe has had the biggest rise in life expectancy at birth. 

GDP has also increased in all these countries. China has had the biggest rise in GDP. 

Life expectancy and GDP are somewhat related. The higher the GDP, the higher the life expectancy, but there are some occurances that don't match this rule and where life expectancy is very high although the GDP is low. 

The average life expectancy of these countries is 72.8. Germany has the highest and Zimbabwe the lowest. 

The distribution of life expectancy is interesting. The most frequent ages are from 75 - 80, but there are also some occurences with low life expectancy.
